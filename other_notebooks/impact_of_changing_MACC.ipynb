```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Abatement

```python
## calculate abatement given a carbon price

def abatement(P, P0, P_h=44, r=20, s=18, Am=1.1):
    A = Am/(1+((P-P0)/P_h)**(-r/s))
    
    return A
```

```python
## calculate abatement given a carbon price
## changes to MAC curve formula

def abatement2(P, P0, P_h=44, z=1, Am=1.1):
    A = Am/(1+((P-P0)/(P_h-P0))**(-z))
    
    return A
```

```python
## calculate abatement given a carbon price
## changes to MAC curve formula

def abatement3(P, P0, P_50=100, r=0.05, s=0.05, P_100=500):
    Am = 1+((P_100-P0)/(P_50-P0))**(-s/r)
    A = Am/(1+((P-P0)/(P_50-P0))**(-s/r))
    
    return A
```

Definite meaning:
- $A_{m}$ – Maximum abatement rate
- $P_{0}$ - Carbon price at initial time

"Proxy for":
- $P_{h}$ - Carbon price at 50% abatement
- $r$ - Timescale of increase in carbon price 
- $s$ – Abatement transition timescale


## Impact of $P_{0} \neq 0$

```python
SCC_0 = np.arange(0, 1000)
plt.plot(abatement(SCC_0, P0=0), SCC_0, label="0")
SCC_25 = np.arange(25, 1000)
plt.plot(abatement(SCC_25, P0=25), SCC_25, label="25")
plt.xlabel("Abatement")
plt.ylabel("Carbon Price")
plt.yscale("log")
plt.legend(title="P0 ($)")
```

## Varying Individual Parameters

```python
P0 = 25  # choose an initial carbon price of $25
SCC_example = np.arange(P0, 1000)
```

### s

```python
s_array = np.arange(10, 40, 5)

for i in range(len(s_array)):
    plt.plot(abatement(SCC_example, P0=P0, s=s_array[i]), SCC_example, label=s_array[i])

plt.xlabel("Abatement")
plt.ylabel("Carbon Price")
plt.yscale("log")
# plt.ylim(0, 100)
plt.legend(title="s")
```

### r

r will obviously have the opposite effect of s (however not really meaningful to vary regarding technological learning whereas s is)

```python
r_array = np.arange(10, 40, 5)

for i in range(len(s_array)):
    plt.plot(abatement(SCC_example, P0=P0, r=r_array[i]), SCC_example, label=s_array[i])

plt.xlabel("Abatement")
plt.ylabel("Carbon Price")
plt.yscale("log")
# plt.ylim(0, 100)
plt.legend(title="r")
```

### P_h

```python
P_h_array = np.arange(20, 100, 10)

for i in range(len(P_h_array)):
    plt.plot(abatement(SCC_example, P0=P0, P_h=P_h_array[i]), SCC_example, label=P_h_array[i])

plt.xlabel("Abatement")
plt.ylabel("Carbon Price")
# plt.yscale("log")
plt.legend(title="P_h")
```

### A_m

```python
Am_array = np.arange(1, 1.5, 0.1)

for i in range(len(Am_array)):
    plt.plot(abatement(SCC_example, P0=P0, Am=Am_array[i]), SCC_example, label=Am_array[i])

plt.xlabel("Abatement")
plt.ylabel("Carbon Price")
plt.yscale("log")
plt.legend(title="Am")
```
