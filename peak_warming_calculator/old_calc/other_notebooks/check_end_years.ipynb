```python
end_year_array = np.linspace(2500, 3500, 10)
```

```python
peak_T_varying_end_year = []

for i in range(len(end_year_array)):
    peak_T_varying_end_year.append(peak_warming_calculator(end_year=end_year_array[i]))
```

## varying final perturbed year (how far out until the SCC is achieved by the final year)

```python
perturbation_year_array = np.linspace(2100, 2500, 10)
```

```python
peak_T_varying_perturbation_year = []

for i in range(len(perturbation_year_array)):
    peak_T_varying_perturbation_year.append(peak_warming_calculator(last_perturbed_year=perturbation_year_array[i], end_year=3500))
```

```python
plt.plot(perturbation_year_array, peak_T_varying_perturbation_year)
plt.xlabel("final perturbed year")
plt.ylabel("peak T")
plt.title("g=0.02")
```

## for lower consumption growth rate

```python
peak_T_varying_perturbation_year_g = []

for i in range(len(perturbation_year_array)):
    peak_T_varying_perturbation_year_g.append(peak_warming_calculator(last_perturbed_year=perturbation_year_array[i], end_year=3500, consumption_growth=0.01))
```

```python
plt.plot(perturbation_year_array, peak_T_varying_perturbation_year_g)
plt.xlabel("final perturbed year")
plt.ylabel("peak T")
plt.title("g=0.01")
```
